```
import matplotlib.pyplot as plt
```

```
x_values = [2014,2015,2016,2017,2018,2019]
y_values = [150,165,170,175,180,190]


plt.style.use('ggplot')
fig,ax = plt.subplots(figsize=(15,9))
ax.plot(x_values, y_values, linewidth=3)

# set chart title and label axes
ax.set_title('The Internal Rate of Return')
ax.set_xlabel('Years',fontsize=14)
ax.set_ylabel('Sales in Millions of $')

# Set size of tick labels
ax.tick_params(axis='both', labelsize = 14)

plt.show()
```

```
window_size = 3

i =0
moving_averages = []
while i < len(y_values) - window_size + 1:
    this_window = y_values[i : i + window_size]

    window_average = sum(this_window) / window_size
    moving_averages.append(window_average)

    i += 1

print(moving_averages)
```

```
time_periods = [1,2,3,4,5]
cash_flow_one = [18,18.9,20.79,22.869,25.1559]
cash_flow_two = [12.5, 18.2, 18.2, 5.2, 5.2]

# bar plot
plt.figure(figsize=(10,9))
plt.bar(time_periods, cash_flow_one, align='center',alpha=0.8)
plt.xlabel('Periods')
plt.ylabel('Cashflow')
plt.title('£ 78.84')

plt.show()
```

```
plt.figure(figsize=(12,9))
plt.bar(time_periods, cash_flow_two,color='blue',align='center',alpha=0.6)
plt.xlabel('Periods')
plt.ylabel('Cashflow')
plt.title('£ 46.84')

plt.show()
```
